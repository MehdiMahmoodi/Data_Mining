## 3- Defining Dummy Variables

### Why dummies:

    .Non-linear realtionships
        .A subset of a variable (a certain range of values) may affect dependent differently, but variable used as a continuous one may not be significant.
    . Easier to interpret
        .for business applications
    . Can handle special cases easily 
        .no record, inquiries only, missing, etc. for credit bureau data, based on dependent variable characteristics for those categories.
        
### How to define them:

    .Compute G/B Percent ratio (Good Column Percent / Bad Column Percent for each category).
    .Use the pattern of these ratios to determine how many categories (and hence number of dummies) to create.
    .Must have a neutral category.
    
### Guidelines:

    .Eg: Ratios get better with age â€“ does that make sense? Why or why not?
    .If a higher age category has lower ratio then combine it with the previous (or next) category. 
    .If pattern is entiresly contrary to business expectation, investigate data, and/or drop the variable.
    .If no pattern (variation in ratios) at all, drop the variable â€“ it has no discriminatory power.
    
### Special cases:

    .Make them Neutral by default.
    .Look at Good/Bad ratio for those categories.
    .Find bucket of legitimate values of variable with similar G/B ratio, and make that the Neutral category also.
    .Assess the impact of special cases on dependent with a separate variable that simply identifies a record as a special case or not.
